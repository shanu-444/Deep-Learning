## Data Preprocessing

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
for column in df:
    print(column, df[column].unique())
```

```python
# Replacing 'No Internet Service' to 'No'

df.replace('No internet service', 'No', inplace=True)
df.replace('No phone service', 'No', inplace=True)

replace_column_values = ['PhoneService', 'Partner', 'Dependents', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                      'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

df.replace({'Male': 0, 'Female': 1}, inplace=True)
df.replace({'No': 0, 'Yes': 1}, inplace=True)
```

```python
# Converting into float

df['TotalCharges']  = pd.to_numeric(df.TotalCharges, errors = 'coerce')
```

#  Using One Hot Encoding

```python
df = pd.get_dummies(data=df, columns = ['Contract', 'InternetService', 'PaymentMethod'])
df.head()
```

## Using MinMaxScaler

```python
scaler_columns = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[scaler_columns] = scaler.fit_transform(df[scaler_columns])
```

# Splitting Training & Test Dataset

```python
x = df.drop('Churn', axis=1)
y = df.Churn

print(x.shape, y.shape)
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split

(X_train, X_test, y_train, y_test) = train_test_split(x, y, test_size=0.2, random_state=10) 
```

# Building The ANN

```python
from tensorflow import keras


ann = tf.keras.Sequential()

# Adding the layers

ann.add(keras.layers.Dense(units=20, activation = 'relu'))
ann.add(keras.layers.Dense(units= 15, activation= 'relu'))

# Output Layer

ann.add(tf.keras.layers.Dense(units=1, activation= 'sigmoid'))

#Compiler

ann.compile(optimizer = 'adam',
           loss = 'binary_crossentropy',
           metrics=['acccuracy'])
```

```python
df = df.astype(np.float64)
```

```python
ann.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=['accuracy'] )

ann.fit(X_train, y_train, epochs=50)
```

```python
ann.evaluate(X_test, y_test)
```
